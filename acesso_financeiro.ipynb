```python
import pandas as pd

# Ler o arquivo, ignorando linhas iniciais irrelevantes
df_idade = pd.read_excel('./Dataset/acesso_financeiro.xlsx', skiprows=3)

#  Remover colunas totalmente vazias
df_idade = df_idade.dropna(axis=1, how='all')

#  Remover linhas totalmente vazias
df_idade = df_idade.dropna(how='all')

#  Redefinir o índice para evitar confusão
df_idade = df_idade.reset_index(drop=True)

# 5 Verificar a primeira linha — se ela contém nomes redundantes, descartá‑la:
if df_idade.iloc[0].isnull().sum() < len(df_idade.columns) / 2:
    # Detecta se a primeira linha tem strings como "Moradia", "Educação" etc.
    if any(isinstance(v, str) for v in df_idade.iloc[0]):
        df_idade = df_idade.drop(0).reset_index(drop=True)

# Renomear colunas com base nos nomes desejados
df_idade.columns = [
    'Unidade_Federacao',
    'Proporcao_Residencia',
    'Proporcao_Vulnerabilidade',
    'IVM_NM',
    'Moradia',
    'Servicos_Publicos',
    'Saude_Alimentacao',
    'Educacao',
    'Servicos_Financeiros',
    'Transporte_Lazer'
]

# Conferir o resultado final
print(df_idade.head())
```
