```python
# Carregamento das bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# Carregamento do dataset de idade
df_idade = pd.read_csv(
    './Dataset/idade.csv',
    sep=';',
    encoding='latin-1',
    skiprows=3,
    quotechar='"',
    low_memory=False
)
```

```python
# Garantir que a coluna nomeia corretamente
df_idade.rename(columns={'Município': 'MUNICIPIO'}, inplace=True)

# Remover números e espaços extras no início do nome
df_idade['MUNICIPIO'] = (
    df_idade['MUNICIPIO']
    .astype(str)
    .str.replace(r'^\d+\s*', '', regex=True)  # remove números no início
    .str.strip()                              # tira espaços extras
)

# Conferir
df_idade.head()
```

```python
# Verificar o número de linhas no DataFrame
len(df_idade)
```

```python
# Verificar número de municípios únicos
df_idade['MUNICIPIO'].nunique()
df_idade[['MUNICIPIO']].head(10)
```

```python
# Salvar arquivo tratado 
df_idade.to_csv('./Dataset/idade_tratado.csv', sep=';', index=False, encoding='utf-8')
```
